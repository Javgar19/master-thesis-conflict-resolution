```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./results/simulations.csv')
```

```python
df.columns
```

```python
def confs_vs(variable, df):
    
    df = df.groupby(variable).sum()
    return df["number_conflicts"]

dss = confs_vs("strength", df)
plt.plot(dss)
```

```python
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(10, 8)
strength_conf = confs_vs("strength", df)
clustering_conf = confs_vs("clustering_coeff", df)
nn_conf = confs_vs("nn_degree", df)
edgedens_conf = confs_vs("edge_density", df)
ax[0, 0].plot(strength_conf)
ax[0, 0].set_xlabel("strength")
ax[0, 0].set_ylabel("number of conflicts")


ax[0, 1].plot(clustering_conf)
ax[0, 1].set_xlabel("clustering_coeff")

ax[1, 0].plot(nn_conf)
ax[1, 0].set_xlabel("nn_degree")
ax[1, 0].set_ylabel("number of conflicts")

ax[1, 1].plot(edgedens_conf)
ax[1, 1].set_xlabel("edge_density")

plt.show()
```
